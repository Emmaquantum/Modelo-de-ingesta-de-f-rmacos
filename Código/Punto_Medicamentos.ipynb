```python
import numpy as np
import matplotlib.pyplot as plt
import random
u = lambda t: np.piecewise(t,t>0,[1,0])
```

# Modelo 1

```python
#Tiempo en Horas del tratamiento
t = 24

ti = np.arange(0,t+0.01,0.1)
# perido de dosificacion
f = 4

#Numero de docis programdas en relacion con el tiempo de trtameinto
n = int(np.floor(t/f))
#n = 3

# rango de la tasa de eliminacion
k = random.uniform(0.173,0.693)

```

```python
r = 0
for i in range(n+1):
    ui = 500*u(ti-i*f)*np.e**(-k*(ti-i*f))
    r += ui
```

```python
plt.plot(ti,r, label=f'Numero de docis {int(t/f)}\n  Tasa de eliminacion {np.round(k,3)}')
plt.grid()
plt.ylabel("Medicamento en mg")
plt.xlabel("Tiempo en horas")
plt.legend()

plt.show()
```

# Modelo 2

```python
# docis para el medelo 2
n2 = 1

# rango de la tasa adsorcion del medicamento
a = random.uniform(0.173,0.693)

#Tasa de eliminacion del cuerpo modelo 2
k2 = random.uniform(0,np.exp(-a)-1/2)
#k2 = np.log(np.exp(-a)-1/2)

#Docis de medicamento apolicado
p = 500
```

```python
r2 = 0
for i in range(n2):
    u2i = (a*p)/(k-a)*(u(ti-4*i)*np.exp(-a*(ti-4*i))-u(ti-4*i)*np.exp(-k*(ti-4*i)))
    r2 += u2i
```

```python
plt.plot(ti,r2, label = f'Numero de docis {n2}\n  Tasa de eliminacion {np.round(k2,3)}\n Tasa de absorcion {a}')
plt.grid()
plt.ylabel("Medicamento en mg")
plt.xlabel("Tiempo en horas")
plt.legend()
plt.show()
```
